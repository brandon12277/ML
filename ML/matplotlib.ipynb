# MATPLOTLIB

Mostly used for data visualization and used to create stunning plot diagrams

## Line Plot 

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(1,20)
Y = np.arange(11,30)
Z = np.arange(31,50)
plt.plot(X,Y,color="red")
plt.show()
```

### Labelling the sides

```python
plt.title("Line Plot")
plt.xlabel("Test case X")
plt.ylabel("Test case Y")

plt.plot(X,Y,color="red",linewidth=3,linestyle=":")
plt.plot(X,Z,color="blue",linewidth=2,linestyle="-.")
plt.show()
```

## Subplot

```python
plt.title("Line Plot")
plt.xlabel("Test case X")
plt.ylabel("Test case Y")

plt.subplot(2,2,1)
plt.plot(X,Y,color="red",linewidth=3,linestyle=":")
plt.grid(True)
plt.subplot(2,2,2)
plt.plot(X,Z,color="blue",linewidth=2,linestyle="-.")
plt.grid(True)
plt.show()
```

## Bar plot

## Box plot

```python
one = [2,4,65,89,100]
two = [5,7,22,32,12]
three = [67,7,23,100,22]

data = [one,two,three]

plt.boxplot(data)
plt.show()
```
