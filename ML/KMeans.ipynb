```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init = 'k-means++',random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.scatter(range(1,11),wcss,color = "blue")
plt.plot(range(1,11),wcss,color = "red")
plt.title("Elbow Method")
plt.xlabel(" range ")
plt.ylabel(wcss)
plt.show()
```

```python
# thus we will use 5 clusters based on the data plot of elbow method

kmeans = KMeans(n_clusters = 5,init = 'k-means++',random_state = 42)
Y = kmeans.fit_predict(X)
```

```python
plt.scatter(X[Y == 0,0],X[Y == 0,1],color = "blue",label = "0")
plt.scatter(X[Y == 1,0],X[Y == 1,1],color = "green",label = "1")
plt.scatter(X[Y == 2,0],X[Y == 2,1],color = "red",label = "2")
plt.scatter(X[Y == 3,0],X[Y == 3,1],color = "orange",label = "3")
plt.scatter(X[Y == 4,0],X[Y == 4,1],color = "yellow",label = "4")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color = "black" ,s = 100 ,label = "centroid")
plt.legend()
plt.show()
```
