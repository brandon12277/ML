## Mapping to higher dimension
In some cases a dataset cannot be linearly separable in the current dimension.Thus to make it separable we have to map the data we have to map 
the data to a higher dimension and then make it separable.

## Types of Kernel functions :
1. Gaussan RBF kernel
2. Sigmoid Kernel
3. Polynomial Kernel

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
X= dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
print(X_train)
print(X_test)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel="rbf",random_state=0)
svc.fit(X_train,Y_train)
```

## Visualizing

```python
from matplotlib.colors import ListedColormap

x_set,y_set = sc.inverse_transform(X_test),Y_test

X1,X2 = np.meshgrid(np.arange(start = x_set[:,0].min()-10,stop = x_set[:,0].max()+10,step = 0.25),
                    np.arange(start = x_set[:,1].min()-1000,stop = x_set[:,1].max()+1000,step = 0.25))

plt.contour(X1,X2,svc.predict(sc.transform(np.array([X1.ravel(),X2.ravel()]).T)).reshape(1,-1),aplha=0.75,cmap = ListedColormap(("red","blue")))
plt.xlim(X1.min(),X1.max())
plt.xlim(X2.min(),X2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0],X_set[y_set == j,1],color = ListedColormap(("red","blue"))(i),label = j)

plt.title("Kernel SVM")
plt.xlabel("Age")
plt.ylabel("Salary Set")
plt.show()
```
