```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset  = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X_train = st.fit_transform(X_train)
X_test = st.fit_transform(X_test)
print(X_train)
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=0.5,random_state=0)
lr.fit(X_train,Y_train)
print(lr.predict(st.fit_transform([[30,87000]])))
```

```python
plt.subplot(1,2,1)
plt.scatter(st.inverse_transform(X_test)[:,-1],Y_test,color="red")
plt.scatter(st.inverse_transform(X_test)[:,-1],lr.predict(X_test),color="blue")

plt.subplot(1,2,2)
plt.scatter(st.inverse_transform(X_train)[:,-1],Y_train,color="red")
plt.scatter(st.inverse_transform(X_train)[:,-1],lr.predict(X_train),color="blue")
plt.show()
```

## Making the confusion matrix for the model

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(Y_test,lr.predict(X_test))
print(cm)
print(accuracy_score(Y_test,lr.predict(X_test)))
```

1. top left - correct predictions of class 0
2. bottom left - incorrect predictions of class 0
3.  bottom right - correct predictions of class 1
3. top right - incorrect predictions of class 1
