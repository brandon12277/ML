# Hierarchical clustering

Hierarchical clustering is another unsupervised machine learning algorithm, which is used to group the unlabeled datasets into a cluster and also known as hierarchical cluster analysis or HCA.

In this algorithm, we develop the hierarchy of clusters in the form of a tree, and this tree-shaped structure is known as the dendrogram


It is of two type : 
1. Agglomerative
2. Divise

## Agglomerative Hc
In this process we follow the bottom up approach which means we start from the lowest position and then work our way up to 1 cluster.
Thus we first take every point as 1 cluster. We then take the clusters with the closest distance and make in into one cluster.
We then continue to do that and keep making clusters while over all reducing the clusters until we get 1 cluster.

The purpose of this is done so that as we move our way up the memory of each clustering is stored in something called dendograms.

### Dendograms
It is the memory of the hc algo which remembers every step taken
Using a dendogram we can find the optimal number of clusters by finding the largest vertical line which doesnt cross 
the extended horizontal line



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method="ward"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean" ,linkage = "ward")
Y = hc.fit_predict(X)
```

```python
plt.scatter(X[Y == 0,0],X[Y == 0,1],color = "blue",label = "0")
plt.scatter(X[Y == 1,0],X[Y == 1,1],color = "green",label = "1")
plt.scatter(X[Y == 2,0],X[Y == 2,1],color = "red",label = "2")
plt.scatter(X[Y == 3,0],X[Y == 3,1],color = "orange",label = "3")
plt.scatter(X[Y == 4,0],X[Y == 4,1],color = "yellow",label = "4")
# plt.scatter(hc.cluster_centers_[:,0],hc.cluster_centers_[:,1],color = "black" ,s = 100 ,label = "centroid")
plt.legend()
plt.show()
```
